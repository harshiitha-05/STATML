```python
import pandas as pd
a=pd.read_csv('/content/drive/MyDrive/Salary_Data (1).csv')
print(a)
```

```python
import pandas as py
target=a['Salary']
print("target variable is:")
print("",target)
features=a.drop(columns = 'Salary')
print("the features are:")
print("",features)
```

```python
from matplotlib import pyplot as plt
plt.scatter(features,target)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the salary dataset

# Create a linear regression object
lr = LinearRegression()

lr.fit(a[['YearsExperience']], a['Salary'])

# Print the slope and intercept of the line
print('Slope:', lr.coef_)
print('Intercept:', lr.intercept_)

# Plot the data points and the best fit line
plt.scatter(a['YearsExperience'], a['Salary'])
plt.plot(a['YearsExperience'], lr.predict(a[['YearsExperience']]), color='red')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Best Fit Line')
plt.show()
```
